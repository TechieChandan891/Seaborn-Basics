```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = [
    "Achintya", "Gulshan", "Ranajit", "Suryansh", "Rohit Shukla", 
    "Deo Ranjan", "Priya", "Abhishek", "Sapna", "Bishal", 
    "Vaibhav", "Akash", "Gaurav", "Aman", "Harshit", 
    "Nandini", "Shubham", "Mannu", "Prabhash"
]

y = [20, 18, 15, 35, 43, 23, 54, 65, 21, 32, 21, 29, 31, 32, 43, 42, 32, 32, 43]


age_dict = {}
for name, age in zip(x, y):
    if name not in age_dict:
        age_dict[name] = {}
    age_dict[name][age] = age_dict[name].get(age, 0) + 1


matrix = np.zeros((len(x), len(set(y))))
for i, name in enumerate(x):
    for j, age in enumerate(sorted(set(y))):
        matrix[i, j] = age_dict[name].get(age, 0)


sns.heatmap(matrix, xticklabels=sorted(set(y)), yticklabels=x)
plt.title('Heatmap of Age Distribution by Name')
plt.xlabel('Age')
plt.ylabel('Name')
plt.xticks(rotation=90)
plt.show()
```
